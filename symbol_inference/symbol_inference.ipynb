```python
from sympy import *
init_printing(use_latex='mathjax')
x_c, y_c = symbols("x_c y_c") # center of the ellipse
x_local, y_local, a, b = symbols("x_local y_local a b")  # x_local**2 / a**2 + y_local**2/b**2 = 1
theta = symbols("theta") # angle for the ellipse
x, y = symbols("x y")

```

```python
X_local = Matrix([[cos(theta), sin(theta)],[-sin(theta), cos(theta)]]) * Matrix([[x-x_c], [y-y_c]])
x_local = X_local[0]
y_local = X_local[1]
x_local
```

```python
ellipse_local = x_local**2/a**2 + y_local**2/b**2 -1
ellipse_local
```

```python
ellipse_local.subs([(x_local, x), (y_local, y)])
```

```python
ellipse_poly = poly(ellipse_local, [x,y])
ellipse_poly
```

```python
ellipse_coeffs = ellipse_poly.coeffs()
print(ellipse_coeffs[0])
print(ellipse_coeffs[1])
print(ellipse_coeffs[2])
print(ellipse_coeffs[3])
print(ellipse_coeffs[4])
print(ellipse_coeffs[5])

ellipse_coeffs
```
